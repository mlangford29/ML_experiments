# Read Data and Describe Metadata

```python
df = pd.read_csv('lung_data.csv')
list(df.columns)
```

## Metadata Description
This dataset contains several features. Here is the column metadata for each of the features:
- inst: [int, categorical] Institution code
- time: [int] Survival time in days. This is the target variable
- status: [int, binary] censoring status. 1=censored, 2=dead
    - censored refers to when a patient is lost to follow-up, withdraws from the study, or is still alive at the end of the study
    - censored observations can be included when using survival analysis models, but probably need to be excluded when using traditional ML methods
- age: [int] age in years
- sex: [int, binary] 1=Male, 2=Female
- ph.ecog: [int, categorical] ECOG performance score (0=good, 5=dead)
    - Assesses the functional status of the patient. A measure of ability to carry out daily activities
- ph.karno: [int] Karnofsky performance score as rated by physician. (0=dead, 100=normal function)
- pat.karno: [int] Karnofsky performance score as rated by patient. (0=dead, 100=normal function)
- meal.cal: [int] Calories consumed at meals (unclear if this is average calories daily or another measurement)
- wt.loss: [int] Weight loss in last six months

## EDA: Analysis of Missing Data
I noticed in the description of the original dataframe that not all values in all features are filled. For many modeling tasks, potential relationships between missing values and the target variable should be explored (to not introduce imputation bias) and imputation strategies need to be explored

I also want to determine if the distribution of the features where there is a missing value in the row is significantly different from the overall distribution. This can also help inform imputation methods

```python
# creating a mask for finding where there are null values
mask = df.isnull()

# compute fraction of null values for each column
missing_ratio = mask.mean(axis=0)

# sort series by fraction of null values
missing_ratio_sorted = missing_ratio.sort_values(ascending=False)
print(missing_ratio_sorted)
```

```python
# pulling out all rows where there is null data for statistics in the next cell
mask = df.isnull()
has_nan = mask.any(axis=1)
df[has_nan].describe()
```

```python
# Is there a significant difference between rows with missing data and rows with filled data?
t_stat, p_value = ttest_ind(df, df[has_nan], nan_policy='omit')

if (p_value.data < 0.05).any():
    print('Given a p-value of {},'.format(round(min(p_value), 3)))
    print('at least one feature is significantly different for rows with missing data')
else:
    print('Given a p-value of {},'.format(round(min(p_value), 3)))
    print('no feature is significantly different for rows with missing data')

print()

# I also want to find out if there is a link between missing data and the target variable

# recreating boolean mask for null values
mask = df.isnull()

# check which rows contain at least one null value
has_nan = mask.any(axis=1)

# filter rows that have at least one null value
nan_rows = df[has_nan]

# create new DF that excludes rows with missing values
no_nan = df.dropna()

# perform two-sample t-test on "time" column of the two DFs
t_stat, p_value = ttest_ind(nan_rows['time'], no_nan['time'])

# inspect output of t-test
if p_value < 0.05:
    print('Given a p-value of {},'.format(round(p_value, 3)))
    print('there is a statistically significant difference in the "time" feature between the two datasets.')
else:
    print('Given a p-value of {},'.format(round(p_value, 3)))
    print('there is no statistically significant difference in the "time" feature between the two datasets.')
```

### Comments on Missing Data
Because there is no significant difference between observations with missing values and full observations, I am comfortable doing a simple median imputation given the skewed nature of the features.

Also, given the small size of the dataset, ML imputation or MICE might not be effective.

```python
df.fillna(df.median(), inplace=True)
```

# Data Preprocessing
Factoring in the knowledge we gained from the exploratory data analysis steps, we can continue with scaling, cleaning values, and creating dummy variables of certain categorical features

For scaling, some of the columns need to be scaled so that all data can be compared properly. Some models may take a column with a larger range of values as being more important than a column with a smaller range of values.

```python
# select subset of columns to scale
cols_to_scale = ['age', 'ph.ecog', 'ph.karno', 'pat.karno', 'meal.cal', 'wt.loss']

# create scaler object and fit to subset of columns
scaler = preprocessing.MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[cols_to_scale]), columns=cols_to_scale)

# update original DF with scaled values
df.update(df_scaled)

# let's create dummy variables for the categorical variable sex
# replace values in "sex" column with strings
df.replace({'sex': {1: 'male', 2: 'female'}}, inplace=True)

# create dummy variables for "sex" column
df = pd.concat([df, pd.get_dummies(df['sex'])], axis=1)

# drop original "sex" column after dummies created
df.drop('sex', axis=1, inplace=True)

# I'm also going to drop the "inst" column since this shouldn't be a feature we take into account for modeling
df.drop('inst', axis=1, inplace=True)
```

```python
# let's do the same thing for "status" with "censored" if 1 or "dead" if 2
# replace values in "status" column with strings
df_dead = df.copy()
df_dead.replace({'status': {1: 'censored', 2: 'dead'}}, inplace=True)

# replace the original status column with 0 and 1, instead of 1 and 2.
# this will be for use in the survival analysis approach
df.replace({'status': {1: 0, 2: 1}}, inplace=True)
df.rename(columns={'status': 'dead'}, inplace=True)

# create dummy variables for "status" column
df_dead = pd.concat([df_dead, pd.get_dummies(df_dead['status'])], axis=1)

# drop original "status" column
df_dead.drop('status', axis=1, inplace=True)

# now we will create two dataframes from the original:
# one will only consider those whose status is "dead".
# survival analysis can take "censored" into account, but traditional approaches cannot.
# The dataframe with only "dead" as the status will be used for traditional approaches

df_dead = df_dead.loc[df_dead['dead'] == 1]
df_dead.drop(['censored', 'dead'], axis=1, inplace=True)
```

Now that the dataset has been explored, outliers have been considered, missing values have been investigated and imputed, dummy variables have been created for categorical features, and extraneous features have been dropped, we can begin modeling

# Modeling: Survival Analysis
Using survival analysis techniques to try to model how the "time" target variable is related to other features, while also being aware of censored records.

For this analysis, I will use a both a Cox's Proportional Hazards model and a Random Survival Forest on the original feature set. Then I will perform some simple feature engineering (without domain knowledge) and see if we can improve the baseline score of the original Cox's Proportional Hazards model.

## Cox's Proportional Hazards model

```python
# split data into training and testing sets
train_df, test_df = train_test_split(df.drop(['female'], axis=1), test_size=0.2, random_state=42)

# instantiate and fit the model
cph = CoxPHFitter()
cph.fit(train_df, duration_col='time', event_col='dead')

# print out the summary of the model's parameter estimations
cph.print_summary()
```

```python
# check the proportional hazards assumption
cph.check_assumptions(train_df, p_value_threshold=0.05)
```

Looks like we have a failure of the proportional hazards test. This means the hazard ratio for meal.cal is not constant over time, which violates the Cox Proportional Hazards model assumption. We can approach this using stratification, where the baseline hazard function is allowed to differ across levels of meal.cal but the effect (hazard ratio) of other covariates remains same across strata. This will eliminate meal.cal from the results because it will be absorbed into the baseline hazard

```python
# train a stratified CPH model
cph_stratified = CoxPHFitter()
cph_stratified.fit(train_df, duration_col='time', event_col='dead', strata=['meal.cal'])
cph_stratified.print_summary()
```

### Important note (for me) from the documentation
The coefficients in a Cox Proportional Hazards model have a multiplicative effect rather than an additive effect. Coefficient values refer to how many times higher the death rate was for changing this covariate while holding the others constant# Looks like we have a failure of the proportional hazards test

```python
# display partial log-likelihood scores for training/testing data for inspection
print('Average partial log-likelihood score on training data: {}'.format(
    round(cph_stratified.score(train_df), 3)))
print('Average partial log-likelihood score on testing data: {}'.format(
    round(cph_stratified.score(test_df), 3)))
```

```python
cph_stratified.plot()
```

```python
# we can pull predictions out on the test dataset pretty quickly
test_results = pd.DataFrame()
test_results['median'] = cph.predict_median(test_df)
test_results['expectation'] = cph.predict_expectation(test_df)
merged_df = pd.merge(test_results, df['time'], left_index=True, right_index=True)
merged_df.head()
```

Predictions can be pulled out for the median expected value, other percentage values, and for expectation.

Here a few of the predicted values are compared against the true label in the test set. Some predictions are good, but there is considerable noise.

## Random Survival Forest
Unfortunately this model requires some different preprocessing techniques, but the concordance index can be easily extracted across the test set

```python
# random survival forest
# need to structure the array properly to use RSF
y = Surv.from_arrays(df['dead'].astype(bool), df['time'])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.drop(['time', 'dead'], axis=1),
                                                    y,
                                                    test_size=0.2,
                                                    random_state=42)

# Fit the Random Survival Forest model
rsf = RandomSurvivalForest(n_estimators=1000,
                           min_samples_split=5,
                           min_samples_leaf=5,
                           random_state=42)
rsf.fit(X_train, y_train)

# Evaluate the model, using a concordance index
print('Concordance index: {}'.format(rsf.score(X_test, y_test)))
```

## Cox's Proportional Hazards Model with Feature Engineering
Looking into ways we can create some additional features that may be useful for fitting and predicting, including interaction features, binning of age, and simple logarithmic transformations

```python
# we don't need to keep the female column since male is already binary
df.drop(['female'], axis=1, inplace=True)

# Interaction features
df['age_ph.karno_interaction'] = df['age'] * df['ph.karno']
df['age_pat.karno_interaction'] = df['age'] * df['pat.karno']
df['ph.karno_meal.cal_interaction'] = df['ph.karno'] * df['meal.cal']
df['pat.karno_meal.cal_interaction'] = df['pat.karno'] * df['meal.cal']

# Binning
df['age_bin'] = pd.cut(df['age'], bins=[0, 30, 50, 70, 100], labels=['<30', '30-50', '50-70', '>70'])

# One-hot encode the bins
df = pd.get_dummies(df, columns=['age_bin'])

# Logarithmic transformations
df['log_meal.cal'] = np.log(df['meal.cal'] + 1)  # the +1 is to avoid log(0)
df['log_wt.loss'] = np.log(abs(df['wt.loss']) + 1) * np.sign(df['wt.loss'])
```

Collinearity is now an issue with fitting a CPH model. We will analyze collinearity in our engineered feature set using a variance inflation factor to compare features

```python
# calculates VIF for each feature in the dataframe
def calculate_vif(df):
    
    vif_data = pd.DataFrame()
    vif_data["Variable"] = df.columns
    vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    
    return vif_data

# Remove dependent variable(s)
X = df.drop(['time', 'dead'], axis=1, errors='ignore')

# Add an intercept to the model for the VIF calculation
X = add_constant(X)

vif = calculate_vif(X)

print(vif)
```

Severe collinearity is detected in some features like meal.cal and wt.loss. Binned variables ended up having some issues, so we'll get rid of those as well.

```python
# Dropping original meal.cal and wt.loss since we have their log transformations
df.drop(columns=['meal.cal', 'wt.loss'], inplace=True)

# Dropping binned age variables
df.drop(columns=['age_bin_<30', 'age_bin_30-50', 'age_bin_50-70', 'age_bin_>70'], inplace=True)
```

```python
# redo the VIF calculation and see if removing these helped
X = df.drop(['time', 'dead'], axis=1, errors='ignore')
X = add_constant(X)
vif = calculate_vif(X)

print(vif)
```

After the removal of the binned age dummy variables and the original meal.cal and wt.loss variables, the VIF scores of the remaining features are more comfortable.

```python
# repeat fitting and summarizing process, this time with feature engineering
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
cph = CoxPHFitter()
cph.fit(train_df, duration_col='time', event_col='dead')
cph.print_summary()
```

```python
# display partial log-likelihood scores for training/testing data for inspection
print('Average partial log-likelihood score on training data: {}'.format(
    round(cph.score(train_df), 3)))
print('Average partial log-likelihood score on testing data: {}'.format(
    round(cph.score(test_df), 3)))
```

```python
cph.plot()
```

```python
# check the proportional hazards assumption
cph.check_assumptions(train_df, p_value_threshold=0.05)
```

# Model Selection
After fitting a Cox’s proportional hazard model to the original feature set and to a transformed feature set, a Random Survival Forest, and several traditional machine learning models, it appears that there is a near performance tie between the two Cox’s proportional hazard models. 

Comparing the survival methods by concordance index:
- Cox's PH model with original features: 0.67
- Cox's PH model with original features after stratification: 0.68
- Cox's PH model with engineered features: 0.69
- Random Survival Forest with original features: 0.55

Both Cox models outperformed the Random Survival Forest, and the specific feature engineering performed did not increase performance on this metric over the original metrics. *I would decide to select the Cox's proportional hazard model with the original feature set because it performs well and would be simpler to productionize*.

# Comments on Model Deployment

The Cox's proportional hazards model could be deployed on an internal platform as an API endpoint. The trained model can be saved as a .pkl file, and a simple API in Python can be used to spin up an EC2 instance to be able to perform single predicitons or batch transformations. The same process could also be containerized and productionized through an external provider, such as AWS SageMaker.

In the deployed Python API, incoming data would need to be transformed (cleaned, imputed, and scaled) according to how it was performed in the training process before the model can be asked to generate predictions. Issues with consistency between training and prediction phases are common sources of problems in real-world deployments.

This specific model can provide median lifetime predictions, expectations, as well as any percentile a user would like to generate (such as 25th and 75th percentiles). These can all be returned as columns in a batch transformation or as several optional values for an individual prediction.

# Comments on Model Monitoring
The deployed model will need to be monitored over time to make sure the input data does not drift significantly from the model's training dataset. Predictions from the model can also be monitored.

This monitoring can be performed automatically using changepoint detection algorithms, such as PELT. T-tests or other statistical tests can also be performed in batches of a given increment to determine the magnitude and significance of population shift from the original training set.

If the deployed model monitoring system detects the input data has drifted significantly from the model's training data, the training process can be kicked off automatically on fresh data, or the model steward can manually re-train it.

In this specific use case of predicting lifetime of patients, it may be difficult to retrain more accurate models on given the delayed nature of obtaining positive events and the general prevalence of censored data.
